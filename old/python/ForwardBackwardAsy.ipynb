```python
# df = df_tD_o
df = df_tM_o

pp = np.array([0,0,1])

Nk = []
Nr = []
Nn = []

for i in range(len(df)):
# for i in range(10):
    k = np.array([df.iloc[i]['pxNTrue'], df.iloc[i]['pyNTrue'], df.iloc[i]['pzNTrue']])
    k = k / np.sum(k**2) ** 0.5

    yp = np.dot(k, pp)
    rp = (1-yp**2)**0.5
    r = np.sign(yp) * (pp-yp*k) / rp
    
    n = np.cross(k, r)
    
    vLep = np.array([df.iloc[i]['pxLepTrue'], df.iloc[i]['pyLepTrue'], df.iloc[i]['pzLepTrue']])
    
    
    Nk.append(np.dot(vLep, k)/(np.dot(vLep, vLep) * np.dot(k, k)))
    Nr.append(np.dot(vLep, r)/(np.dot(vLep, vLep) * np.dot(r, r)))
    Nn.append(np.dot(vLep, n)/(np.dot(vLep, vLep) * np.dot(n, n)))
    
Nk = np.array(Nk)
Nr = np.array(Nr)
Nn = np.array(Nn)
```

```python
Ck = 2 * (np.sum(Nk>0) - np.sum(Nk<0)) / len(Nk)
Cr = 2 * (np.sum(Nr>0) - np.sum(Nr<0)) / len(Nr)
Cn = 2 * (np.sum(Nn>0) - np.sum(Nn<0)) / len(Nn)
```

```python
CkNum_error = (np.sum(Nk>0) + np.sum(Nk<0))**0.5
CrNum_error = (np.sum(Nr>0) + np.sum(Nr<0))**0.5
CnNum_error = (np.sum(Nn>0) + np.sum(Nn<0))**0.5
```

```python
Ck_error = Ck * ((CkNum_error/(np.sum(Nk>0) - np.sum(Nk<0)))**2 + (len(Nk)**0.5/len(Nk))**2)**0.5
Cr_error = Cr * ((CrNum_error/(np.sum(Nr>0) - np.sum(Nr<0)))**2 + (len(Nr)**0.5/len(Nr))**2)**0.5
Cn_error = Cn * ((CnNum_error/(np.sum(Nn>0) - np.sum(Nn<0)))**2 + (len(Nn)**0.5/len(Nn))**2)**0.5
```
